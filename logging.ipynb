```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('./simulation_log.pickle', 'rb') as f:
    log_data = pickle.load(f)
print(len(log_data))
```

```python
sorted_data = sorted(log_data.items(), key=(lambda x: x[1][1]), reverse=True)
len(sorted_data)
```

```python
for i in range(len(sorted_data)):
    print(len(sorted_data[i][1][0]))
```

```python
for i in range(10):
    data_as_array = np.array(sorted_data[i][1][0]).T
    plt.plot(data_as_array[0], data_as_array[1])
plt.savefig('simulation_path_top10.png')
plt.show()
```

```python
for i in range(len(sorted_data)-1, len(sorted_data)-11, -1):
    data_as_array = np.array(sorted_data[i][1][0]).T
    plt.plot(data_as_array[0], data_as_array[1])
plt.savefig('simulation_path_bottom10.png')
plt.show()
```

```python
for i in range(len(sorted_data)):
    data_as_array = np.array(sorted_data[i][1][0]).T
    plt.plot(data_as_array[0], data_as_array[1])
plt.savefig('simulation_path_all.png')
plt.show()
```
